The goal of this notebook is to short data from both the timed and nontimed conditions. 

Import pandas as pd to start the script and refer to "pandas" as "pd" 

```python
import pandas as pd
```

- Create variabled 'df1' and 'df2' and make it equal to the data file names. We are commanding the pd(pandas) to read the csv file from the computer. 
- Sort the data frame (df) to only include the following headers;'subject','group','blockcode', 'trialcode','response', and 'correct'. 
- Concatenate the files, ignoring for index and sort since both files have alreay had their columns sorted
- Rename the new data frame 'df3'

```python
df1=pd.read_csv('/Users/paulpacheco/Desktop/Test Data Files/timed_raw_2203040054.csv')
df1=df1[['subject','group','blockcode','trialcode','response','correct']]
df2=pd.read_csv('/Users/paulpacheco/Desktop/Test Data Files/nontimed_raw_2203040056.csv')
df2=df2[['subject','group','blockcode','trialcode','response','correct']]
df3= pd.concat([df1, df2], ignore_index=True, sort=False)
df3
```

Variable Names: Subject = Participant ID, Group = condition (1=timed, 2=nontimed), Trialcode = Problem Number, Response = Participant Answer Input

- Remove trial_survey from the blockcode data coloumn
    - This is done by commanding the code to return values from the df file under 'blockcode' that are not equal to 'trial_survey' 

```python
df3=df3[df3['blockcode']!='trial_survey']
df3
```

- We also want to filter out and 'p' strings from the data. 'p' represents the waiting screen in between math expressions, thus it is not needed for data analysis. 
- P needs to be removed from the 'trialcode' data column. This is done by commanding the code to return values from the df file under 'trialcode' that are not equal to 'p'
- To previes all of the data change display setting by using 
    - pd.set_option('display.max_rows', 1000), Allows to see 1000 rows
    - pd.set_option('display.max_column', 10), Allows to see 10 columns

```python
df3=df3[df3['trialcode']!='p']
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_column', 10)
df3
```
