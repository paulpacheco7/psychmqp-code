- Establish a mean series adjusting for coloumn number in the data frame, while allowing you to rename the new variables
- Concatenate the file to include the new varibales once while running the script

```python
def extract_questions_mean(df, low_q, max_q, name):
    question_df = df.iloc[:, low_q+1: max_q+2]
    mean_series = question_df.mean(axis=1).rename(name)
    return pd.concat([df, mean_series], axis=1)

df = extract_questions_mean(df, 1,9, 'mean_amas')
df = extract_questions_mean(df,10,13, 'mean_pme')
df = extract_questions_mean(df,14,14, 'mean_mu')
df = extract_questions_mean(df,15,17, 'mean_mc')
df = extract_questions_mean(df,18,20, 'mean_mv')
df = extract_questions_mean(df,21,24, 'mean_ms')
df
```
